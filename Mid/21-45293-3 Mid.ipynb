## 1) List Operations:

```python
# List of integers containing numbers from 1 to 10.
list_of_integers = [i for i in range(1, 11)]

# Appending the square of each number to the list.
list_of_integers += [i*i for i in range(1, 11)]

# Finding the sum of all elements in the list.
sum_of_list = sum(list_of_integers)

print (list_of_integers)
print (sum_of_list)
```

## 2) Tuple Operations:

```python
# Tuple containing the first five even numbers.
even_numbers_tuple = (2, 4, 6, 8, 10)

# Finding the product of all elements in the tuple.
product_of_evens = even_numbers_tuple[0] * even_numbers_tuple[1] * even_numbers_tuple[2] * even_numbers_tuple[3] * even_numbers_tuple[4]

print (even_numbers_tuple)
print (product_of_evens)
```

## 3) Set Operations:

```python
# Two sets of odd numbers between 1 to 10.
odd_set_1 = {1, 3, 5}
odd_set_2 = {5, 7, 9}

# Finding the union and intersection of the two sets.
odd_set_union = odd_set_1.union(odd_set_2)
odd_set_intersection = odd_set_1.intersection(odd_set_2)

print (odd_set_1)
print (odd_set_2)
print (odd_set_union)
print (odd_set_intersection)
```

## 4) Dictionary Operations:

```python
# A dictionary with the first five lowercase alphabets and their ASCII values.
values = {chr(97 + i): 97 + i for i in range(5)}

print (values)
```

## 5) Functional Programming:

```python
# A function to square a number.
def square(x):
    return x * x

# Using map to apply the square function to each element from 1 to 10.
squared_numbers = list(map(square, range(1, 11)))

print (squared_numbers)
```
